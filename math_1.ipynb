### 3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.


```python
def decart(R, alfa):
    return (R*numpy.cos(alfa), R*numpy.sin(alfa))
print(decart(1,0), decart(1, numpy.pi/2), decart(numpy.sqrt(2), numpy.pi/4))
```

### 4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
x = numpy.setdiff1d(numpy.linspace(-2, 4.5, 1001), numpy.array([0])) 
y1 = x * x - 1
y2 = 1-(1-numpy.exp(x))/x # убедимся, что для x=0 y=-1 - тоже решение
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
from scipy.optimize import fsolve

def equations(eq):
    x, y = eq
    return (y - x*x +1, numpy.exp(x) + x*(1-y) - 1)
print(fsolve(equations, (-2, 0)), 
      fsolve(equations, (0, -1)), #не забудем потерянный корень
      fsolve(equations, (2.5, 0)), 
      fsolve(equations, (5, 0)), sep='\n')
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1


```python
def f(x):
    return numpy.exp(x) + 2*x - x**3 - 1

x = numpy.linspace(-2, 4.5, 1001)
x_solve = x[f(x) > 0]
plt.scatter(x_solve, f(x_solve));
print(f"x \u2208 (-\u221e, {fsolve(f, -1.5)[0]})", 
      f"({fsolve(f, 0)[0]}, {fsolve(f, 2.5)[0]})", 
      f"({fsolve(f, 4)[0]}, +\u221e)", sep='\u222a')
```
